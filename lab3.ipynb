```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(loc = 0 , scale =1 , size = 1000)
```

```python
plt.hist(data,bins=30 , density= True)
```

```python
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# power law distribution
data = np.random.power(a=5 , size = 10000)

plt.hist(data , bins = 50 , density = True)
plt.title("power law distribution")
plt.xlabel("value")
plt.ylabel("Probability Density")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# ----------------------------------------------------
# 1. Create one dataset
# ----------------------------------------------------
np.random.seed(42)

data = pd.DataFrame({
    "Age": np.random.randint(20, 60, 200),
    "Salary": np.random.normal(50000, 10000, 200),
    "Department": np.random.choice(["HR", "IT", "Finance", "Marketing"], 200)
})

print("Dataset Head:")
print(data.head())

# ----------------------------------------------------
# 2. Histogram of Age (Empirical Distribution)
# ----------------------------------------------------
plt.hist(data["Age"], bins=15, edgecolor='black')
plt.title("Histogram of Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()



```

```python
# ----------------------------------------------------
# 3. Bar Plot (Department Counts)
# ----------------------------------------------------
dept_counts = data["Department"].value_counts()

plt.bar(dept_counts.index, dept_counts.values)
plt.title("Employees per Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()

#
```

```python
# ----------------------------------------------------
# 4. Normal Distribution (Salary)
# ----------------------------------------------------
mu, sigma = data["Salary"].mean(), data["Salary"].std()

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = norm.pdf(x, mu, sigma)

plt.hist(data["Salary"], bins=20, density=True, alpha=0.6, edgecolor='black')
plt.plot(x, y)
plt.title("Normal Distribution of Salary")
plt.xlabel("Salary")
plt.ylabel("Density")
plt.show()
```
