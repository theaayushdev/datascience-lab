```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10.50)
```

```python
y=np.sin(x)
```

```python
df = pd.DataFrame({
    "x value" : x,
    "sin value " : y
})
print("Data frame head")
print(df.head())
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("X")
plt.ylabel("sin(X)")
plt.show()
```
